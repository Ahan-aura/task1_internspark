```python
data.drop('address',axis=1,inplace=True)
```

```python
data.drop('name',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cat=['online_order','book_table','location','rest_type','cuisines','listed_in(type)']
for col in cat:
  data[col]=le.fit_transform(data[col].astype(str))
```

```python
x=data.drop('rate',axis=1)
y=data['rate']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,r2_score
rf=RandomForestRegressor(n_estimators=300,random_state=42)
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
print(mean_squared_error(y_test,y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
# 1.Ratings
# Most restaurants are rated between 3.5 and 4.2.
# Very few restaurants receive ratings below 2.5 or above 4.8.

 2. Cuisine Trends
# Some cuisines consistently receive higher ratings than others.
# Multi-cuisine restaurants dominate the dataset.

# 3. Location Preferences
# Certain locations contain a much higher concentration of restaurants.
# Popular restaurant hubs also tend to have better average ratings.

# 4. Price vs Rating
# Cost has only a weak positive relationship with ratings.
# Expensive restaurants are not necessarily better rated.

# 5. Customer Engagement
# Restaurants with more votes generally achieve higher ratings, suggesting popularity and visibility contribute to stronger reputations.
```
